```python
#series对象创建,一维数组
data = pd.Series([1,2,3,4,5,6,7,8,9])
```

```python
#获取索引index，数据values
print(data.index)
print(data.values)
```

```python
#指定索引
data = pd.Series([1,2,3,4],index=['a','b','c','d'])
data
```

```python
#是呀list指定index
data = pd.Series([1,2,3,4],index=list('abcd'))
data
```

```python
#传入字典创建
data = {'name':'张三','age':23,'sex':'male'}
data
```

```python
#字典创建，并指定index
pd.Series(data,index=['name','age'])
```

```python
#创建DataFrame对象，传入两个Series对象作为字典的值
p_dict={'name':'张三','age':23,'sex':'male'}
a_dict={'name':'王五','age':27,'sex':'male'}
p_series=pd.Series(p_dict)
a_series=pd.Series(a_dict)
p_series
a_series
#将series对象作为字典的值传入pd.DataFrame
df = pd.DataFrame({'area':a_series,'population':p_series})
df
```

```python
df.index
df.values
df.columns
```
