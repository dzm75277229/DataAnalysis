# ndarray
## 1、ndarray的特性
### 多维性

```python
import numpy as np
```

```python
arr = np.array(5) #创建0维的数组
print(arr)
print('arr的维度：',arr.ndim) # number of dimension
```

```python
arr = np.array([1])  #创建1维的数组
print(arr)
print('arr的维度：', arr.ndim)  # number of dimension
```

```python
arr = np.array([[1,2],[3,4]])  #创建2维的数组
print(arr)
print('arr的维度：', arr.ndim)  # number of dimension
```

### 同质性

## 2、ndarray的属性

```python
arr = np.array(1)
print(arr)
print("数组的形状：",arr.shape)#按ctrl+k提交
print("数组的维度",arr.ndim)
print("元素的个数",arr.size)
print("元素的类型：",arr.dtype)
```

```python
arr = np.array([1,2.5])
print(arr)
print("数组的形状：",arr.shape)
print("数组的维度",arr.ndim)
print("元素的个数",arr.size)
print("元素的类型：",arr.dtype)
```

```python
arr = np.array([[1,2],[3,4]])
print(arr)
print(arr)
print("数组的形状：",arr.shape)#两行两列
print("数组的维度",arr.ndim)
print("元素的个数",arr.size)
print("元素的类型：",arr.dtype)
print("元素的转置",arr.T)
```

## 3、ndarry的创建

```python
#基础的方法arr = np.array
arr = np.array([1,2,3])
print(arr)
```

```python
list1=[1,2,3]
arr=np.array(list1,dtype=np.float64)
print(arr)
print(arr.ndim)
```

```python
#copy
arr1 = np.copy(arr)
print(arr1)
arr1[0]=8
print(arr1)
print(arr)
```

```python
#预定义形状
#全0 全1 未初始化 固定值
#全0
arr = np.zeros((2,1),dtype=np.int64)#两行一列
print(arr)
arr1 = np.zeros((3,),dtype=np.int64)#没有写列，默认是一列
print(arr1)
```

```python
#全1
arr = np.ones((2,1),dtype=np.int64)
print(arr)
```

```python
# 未初始化
arr = np.empty((9,2))
print(arr)
```

```python
# 固定值
arr = np.full((3,4),2026)
print(arr)
```

```python
arr1 = np.zeros_like(arr)
print(arr1)
```

```python
arr1 = np.full_like(arr,2026)
print(arr1)
```

```python
# 等差数列 1 2 3 4 5
arr = np.arange(1,10,3)#start end(not include) step
print(arr)
```

```python
# 等间隔数列
app = np.linspace(1,10,5) #最后一个数表示均匀的分成几份，10也是包含在内的
print(app)
```

```python
# 对数间隔数列
arr = np.logspace(0,10,25,base=2)
print(arr)
```
