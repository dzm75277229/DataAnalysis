# ndarray
## 1、ndarray的特性
### 多维性

```python
import numpy as np
```

```python
arr = np.array(5) #创建0维的数组
print(arr)
print('arr的维度：',arr.ndim) # number of dimension
```

```python
arr = np.array([5]) #创建1维的数组
print(arr)
print('arr的维度：',arr.ndim) # number of dimension
```

```python
arr = np.array([[1,2],[3,4]]) #创建1维的数组
print(arr)
print('arr的维度：',arr.ndim) # number of dimension
```
