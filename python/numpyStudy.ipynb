## 3、ndarry的创建

```python
#基础的方法arr = np.array
arr = np.array([1,2,3])
print(arr)
```

```python
list1=[1,2,3]
arr=np.array(list1,dtype=np.float64)
print(arr)
print(arr.ndim)
```

```python
#copy
arr1 = np.copy(arr)
print(arr1)
arr1[0]=8
print(arr1)
print(arr)
```

```python
#预定义形状
#全0 全1 未初始化 固定值
#全0
arr = np.zeros((2,1),dtype=np.int64)#两行一列
print(arr)
arr1 = np.zeros((3,),dtype=np.int64)#没有写列，默认是一列
print(arr1)
```

```python
#全1
arr = np.ones((2,1),dtype=np.int64)
print(arr)
```

```python
# 未初始化
arr = np.empty((9,2))
print(arr)
```

```python
# 固定值
arr = np.full((3,4),2026)
print(arr)
```

```python
arr1 = np.zeros_like(arr)
print(arr1)
```

```python
arr1 = np.full_like(arr,2026)
print(arr1)
```

```python
# 等差数列 1 2 3 4 5
arr = np.arange(1,10,3)#start end(not include) step
print(arr)
```

```python
# 等间隔数列
app = np.linspace(1,10,5) #最后一个数表示均匀的分成几份，10也是包含在内的
print(app)
```

```python
# 对数间隔数列
arr = np.logspace(0,10,25,base=2)
print(arr)
```

```python
# 特殊矩阵
# 单位矩阵，对称矩阵，转置矩阵
arr=np.eye(2,dtype=int)
print(arr)
```

```python
# 对角矩阵
arr = np.diag([1,2,3,4])
print(arr)
```

```python
# 随机数组的生成
# 生成0-1之间随机的浮点数
arr = np.random.rand(3,4)#生成三行四列的数据
print(arr)
```

```python
# 生成指定范围区间的随机浮点数
arr = np.random.uniform(low=-1,high=1,size=(3,4))
print(arr)
```

```python
# 生成指定范围的随机证书
arr = np.random.randint(1,10,size=(3,4))
print(arr)
```

```python
# 生成随机数列（正态分布）
arr = np.random.randn(2,3)
print(arr)
```

## 5、索引与切片


```python
# 一维数组的索引与切片
arr = np.random.randint(1,10,size=(3,4))
print(arr)
print(arr[0])
print(arr[1:2])#不包含2
print(arr[(arr<10) &(arr>6)])
print(arr[slice(0,2)])#不包含2
```

# 6、ndarray的运算

```python
# 算数运算
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a+b)
```

```python
c = np.array([1,2,3])
for i in range(len(c)):
    print(c[i])
```

```python
a = np.array([1,2,3])
print(a+3)
```

```python
# 广播机制，其中一个维度是1，另一个维度相同
a = np.array([1,2,3])#1*3
b = np.array([[1],[2],[3]]) #3*1
print(a+b)
'''纬度不同,进行扩展
a
1 2 3
1 2 3
1 2 3
b
1 1 1
2 2 2
3 3 3
'''
```

```python
# 矩阵运算,需要形状相同或满足广播特性，形状相同的对应位置相乘
a = np.array([[1,2,3],[3,2,1]])
b = np.array([[2,2,2],[1,2,1]])
print(a*b)
c = np.array([[1,1,1],[1,2,1],[0,1,1]])
d = np.array([[1,1,1],[1,2,1],[0,1,1]])
print(c*d)
'''
a
1 2 3
3 2 1
b
2 2 2
1 1 1
'''
```

### 2、统计函数
求和、计算、计算中位数、标准差、方差、查找最大值、查找最小值、计算分位数、累积和、累计差

```python
arr = np.random.randint(1,10,8)
print(arr)
#求和
print(np.sum(arr))
#计算平均值
print((np.average(arr)))
#计算最大值
print(np.max(arr))
#计算最小值
print(np.min(arr))
#计算中位数,奇数：排序后中间数，偶数：排序后中间两个数的平均值
print(np.median(arr))
#计算标准差、方差
print(np.var(arr))
print(np.std(arr))
```
