# ndarray
## 1、ndarray的特性
### 多维性

```python
import numpy as np
```

```python
arr = np.array(5) #创建0维的数组
print(arr)
print('arr的维度：',arr.ndim) # number of dimension
```

```python
arr = np.array([1])  #创建1维的数组
print(arr)
print('arr的维度：', arr.ndim)  # number of dimension
```

```python
arr = np.array([[1,2],[3,4]])  #创建2维的数组
print(arr)
print('arr的维度：', arr.ndim)  # number of dimension
```

### 同质性

```python
arr = np.array([1,'hello'])  #不同的数据元素成员
print(arr)
```

## 2、ndarray的属性

```python
arr = np.array(1)
print(arr)
print("数组的形状：",arr.shape)
```
